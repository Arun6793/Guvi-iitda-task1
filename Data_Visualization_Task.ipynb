```python
import pandas as pd
data = pd.read_excel("/content/coronacases.xlsx")
```

```python
df=pd.DataFrame(data)
df
```

```python
df['Date'],df['cases']
df['Date'],df['active cases']
df['Date'],df['deaths']
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
fig=plt.figure()
s1=plt.subplot2grid((3,5),(0,1),colspan = 3)
s2=plt.subplot2grid((3,5),(1,0),colspan=2,rowspan=2)
s3=plt.subplot2grid((3,5),(1,3),colspan=2,rowspan=2)

s1.plot(df['Date'],df['cases'])
plt.xlabel('Date')
plt.ylabel('cases')
s2.plot(df['Date'],df['active cases'])
s3.plot(df['Date'],df['deaths'])
plt.xticks(rotation=90)
plt.tight_layout()
```
